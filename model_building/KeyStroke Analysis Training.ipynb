```python
df = pd.read_csv('keystrokes_dynamics.csv')
```

```python
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
y = y.reshape(-1, 1)
y = y-1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

from tensorflow.keras.utils import to_categorical

y_train_encoded = to_categorical(y_train, num_classes=unique_cols)
y_test_encoded = to_categorical(y_test, num_classes=unique_cols)


unique_cols = df['user'].nunique()

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential()
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=unique_cols, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train_encoded, epochs=1000, batch_size=32, validation_data=(X_test, y_test_encoded))
```
